```python
import pandas as pd
import numpy as np

import seaborn as sns
%matplotlib inline

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#load the dataset
cosmetics_df = pd.read_csv("D:\MLRS\Data Mining & Predictive Analytics_Classes_2017-18\DMPA_Data for association rule\Cosmetics.csv")
cosmetics_df.head()
```

```python
cosmetics_df.drop('Trans. ',inplace = True, axis = 1)
```

```python
frequent_itemsets = apriori(cosmetics_df, min_support=0.05, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules.head()
```

```python
#load the dataset
retail_df = pd.read_csv("D:\MLRS\Data Mining & Predictive Analytics_Classes_2017-18\DMPA_Data for association rule\market basket.csv")
retail_df.head()
```

```python
retail_df.drop(['cardid','value','pmethod','sex','homeown','income','age'],inplace = True, axis = 1)
```

```python
for i in retail_df.columns:
    retail_df[i] = retail_df[i].apply(lambda x: 0 if (x == 'F') else 1 )
```

```python
frequent_itemsets = apriori(retail_df, min_support=0.05, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules.head()
```
